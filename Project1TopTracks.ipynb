```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth
import pandas as pd

#Authentification 
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id="bee7431f4af34a9299881434e09a09cf",
    client_secret="bfbe91a137f24b02b78f0818c1989205",
    redirect_uri="http://127.0.0.1:8888/callback",
    scope="user-top-read"
))
```

```python
# Function to get top tracks 
def get_top_tracks(time_range="short_term", limit=50):
    results = sp.current_user_top_tracks(time_range=time_range, limit=limit)
    tracks = []
    for idx, item in enumerate(results['items']):
        track = {
            "rank": idx + 1,
            "name": item['name'],
            "artist": item['artists'][0]['name'],
            "track_id": item['id'],
            "url": item['external_urls']['spotify']
        }
        tracks.append(track)
    return pd.DataFrame(tracks)
```

```python
# Top tracks for each time range
df_top_month = get_top_tracks("short_term")      # Top 50 songs from past month
df_top_6months = get_top_tracks("medium_term")   # Top 50 songs from past 6 months
df_top_alltime = get_top_tracks("long_term")     # Top 50 songs from previous years
```
